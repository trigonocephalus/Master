```python
import numpy as np
from matplotlib import pylab as plt
```

```python
x = np.linspace(-10, 10, 50)
x
```

```python
y = 2.3 * x - 3
y
```

```python
a = 5.3
b = -5.1
delta = np.inf
eta = 0.0001
D = [(a, b)]
```

```python
while delta > 0.0001:
    hy = a * x + b
    e = (y - hy)
    a = a + 2 * eta * (e * x).sum()
    b = b + 2 * eta * e.sum()
    D.append((a, b))
    delta = np.fabs(np.array(D[-1]) - np.array(D[-2])).mean()
```
